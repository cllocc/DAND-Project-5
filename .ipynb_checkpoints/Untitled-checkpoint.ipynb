```python
#Query our data and store it into a pandas dataframe
gender = sql.read_sql('SELECT member_gender, COUNT(member_gender) as count FROM bike_share WHERE member_gender !="" GROUP BY member_gender ORDER BY COUNT(member_gender) DESC', con)
gender
```

```python
#create our bar chart
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="member_gender", y="count", data=gender,label="Total", color="#79B8D5")

plt.xlabel('')
plt.ylabel('Count')
plt.title("Bike Share User Gender")
#plt.savefig('name_here')
```

### Lets create user groups based on generation.
<p>
<li><em>Baby Boomers</em> between years 1946 – 1964</li>
    <li><em>Generation X</em> between years 1965 – 1980</li>
    <li><em>Millennials</em> between years 1981 – 1996</li>

```python
#Query Boomers
boomers = df[(df['member_birth_year'] >= 1946) & (df['member_birth_year'] <= 1964)]
#Query Genx
genx = df[(df['member_birth_year'] >= 1965) & (df['member_birth_year'] <= 1980)]
#Query Milennials
mil = df[(df['member_birth_year'] >= 1981) & (df['member_birth_year'] <= 1996)]
```

```python
crs = {'init': 'epsg:4326'}
```

```python
geometry = [Point(xy) for xy in zip(df["end_station_longitude"], df["end_station_latitude"])]
geometry[:3]
```

```python
geo_df = gdp.GeoDataFrame(df,#Specify our data
                         crs = crs,#specify our coordinate reference system
                         geometry = geometry)#Specify the geometry list we created
geo_df.head()
```

```python
mil = geo_df[(geo_df['member_birth_year'] >= 1981) & (df['member_birth_year'] <= 1996)]
#Query Boomers
boomers = geo_df[(geo_df['member_birth_year'] >= 1946) & (df['member_birth_year'] <= 1964)]
#Query Genx
genx = geo_df[(geo_df['member_birth_year'] >= 1965) & (df['member_birth_year'] <= 1980)]
```
